```python
print(df.columns.tolist())
```

```python
df.rename(columns={'default.payment.next.month': 'default'}, inplace=True)
```

```python
sns.countplot(x='EDUCATION', hue='default', data=df)
plt.title("Education vs Default")
plt.show()
```

```python
sns.countplot(x='MARRIAGE', hue='default', data=df)
plt.title("Marriage vs Default")
plt.show()
```

```python
sns.histplot(df['LIMIT_BAL'], kde=True, bins=30)
plt.title("Distribution of Credit Limit")
plt.xlabel("Limit Balance")
plt.show()
```

```python
sns.boxplot(x='default', y='LIMIT_BAL', data=df)
plt.title("Credit Limit by Default Status")
plt.show()
```

```python
sns.histplot(df['AGE'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

sns.boxplot(x='default', y='AGE', data=df)
plt.title("Age vs Default")
plt.show()
```

```python
pay_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

for col in pay_cols:
    sns.countplot(x=col, hue='default', data=df)
    plt.title(f"{col} vs Default")
    plt.show()
```

```python
bill_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

for col in bill_cols:
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()

    sns.boxplot(x='default', y=col, data=df)
    plt.title(f"{col} by Default")
    plt.show()
```

```python
pay_amt_cols = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

for col in pay_amt_cols:
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()

    sns.boxplot(x='default', y=col, data=df)
    plt.title(f"{col} by Default")
    plt.show()
```

```python
import numpy as np

plt.figure(figsize=(14, 10))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['default'].value_counts().plot.pie(labels=["No Default", "Default"], autopct='%1.1f%%', startangle=90)
plt.title("Default Payment Distribution")
plt.ylabel("")
plt.show()
```
